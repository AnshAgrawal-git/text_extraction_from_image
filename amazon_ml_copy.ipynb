```python
from PIL import Image, ImageEnhance, ImageFilter, ImageFile
import numpy as np
import cv2
import os
import pytesseract 
# Allow loading of truncated images
ImageFile.LOAD_TRUNCATED_IMAGES = True

# Tesseract OCR path (update this according to your environment)
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# Source and destination directories
source_dir = "C:/Users/lenovo/Desktop/ML/amazon_ml/image_test"
destination_dir = "C:/Users/lenovo/Desktop/ML/amazon_ml/processed_image_test"

# Ensure destination directory exists
if not os.path.exists(destination_dir):
    os.makedirs(destination_dir)

# Process each image in the source folder
for filename in os.listdir(source_dir):
    if filename.endswith((".jpg", ".jpeg", ".png")):  # Process only image files
        image_path = os.path.join(source_dir, filename)
        
        try:
            # Load the image with Pillow
            image = Image.open(image_path)

            # Convert the image to grayscale
            gray_image = image.convert("L")

            # Enhance contrast
            enhancer = ImageEnhance.Contrast(gray_image)
            contrast_factor = 1.5  # Increase contrast by 50%
            enhanced_image = enhancer.enhance(contrast_factor)

            # Convert to NumPy array for OpenCV processing
            enhanced_array = np.array(enhanced_image)

            # Apply CLAHE (Contrast Limited Adaptive Histogram Equalization)
            clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
            clahe_image_array = clahe.apply(enhanced_array)

            # Convert the CLAHE-enhanced image back to a PIL image
            clahe_image = Image.fromarray(clahe_image_array)

            # Resize the image
            target_height = 2000
            aspect_ratio = clahe_image.width / clahe_image.height
            new_width = int(target_height * aspect_ratio)
            new_size = (new_width, target_height)
            resized_image = clahe_image.resize(new_size, Image.Resampling.LANCZOS)

            # Apply sharpening
            sharpened_image = resized_image.filter(ImageFilter.SHARPEN)

            # Apply Unsharp Masking
            unsharp_image = sharpened_image.filter(ImageFilter.UnsharpMask(radius=2, percent=150, threshold=3))

            # Reduce the size after sharpening if needed (optional)
            reduced_size = (int(unsharp_image.width * 0.5), int(unsharp_image.height * 0.5))
            reduced_image = unsharp_image.resize(reduced_size, Image.Resampling.LANCZOS)

            # Save the processed image to the destination directory
            output_path = os.path.join(destination_dir, f"processed_{filename}")
            reduced_image.save(output_path)

            print(f"Processed and saved: {output_path}")

#Perform OCR using Tesseract to extract text
            
            ocr_result = pytesseract.image_to_string(reduced_image)
            print(f"OCR Result for {filename}: {ocr_result}")

            print(f"Processed and saved: {output_path}")

        except Exception as e:
            print(f"Error processing {filename}: {e}")

print("Image processing and OCR extraction completed for all images.")        
```
